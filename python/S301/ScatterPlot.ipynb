```python
import matplotlib.pyplot as plt
```

```python
# Sample data
advertising_spending = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]  # Advertising spending in thousands
sales_revenue = [25, 30, 35, 40, 50, 55, 60, 70, 75, 85]        # Sales revenue in thousands
```

```python
# Create a scatter plot
plt.scatter(advertising_spending, sales_revenue, color='blue', marker='o')
```

```python
# Add titles and labels
plt.title('Advertising Spending vs. Sales Revenue')
plt.xlabel('Advertising Spending (in thousands)')
plt.ylabel('Sales Revenue (in thousands)')
```

```python
# Display the plot
plt.grid(True)  # Optional: add grid lines for better readability
plt.show()
```
