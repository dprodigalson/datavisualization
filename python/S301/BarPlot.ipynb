```python
import matplotlib.pyplot as plt
```

```python
# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
revenue = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75]  # Revenue in thousands
```

```python
# Create a bar plot
plt.bar(months, revenue, color='skyblue')
```

```python
# Add titles and labels
plt.title('Monthly Sales Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue (in thousands)')
```

```python
# Display the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: add horizontal grid lines
plt.show()
```
