```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Create a DataFrame with example data
data = {
    '8 AM': [50, 60, 55, 70, 65, 80, 90],
    '12 PM': [80, 85, 90, 95, 100, 110, 120],
    '4 PM': [70, 75, 80, 85, 90, 95, 100],
    '8 PM': [40, 45, 50, 55, 60, 65, 70]
}
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
# Convert data to DataFrame
df = pd.DataFrame(data, index=days)
```

```python
# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='YlGnBu', fmt='d')
```

```python
# Add titles and labels
plt.title('Customer Purchases Heatmap')
plt.xlabel('Time of Day')
plt.ylabel('Day of the Week')
```

```python
# Display the plot
plt.show()
```
